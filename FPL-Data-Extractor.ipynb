```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('data/cleaned_merged_seasons.csv',index_col=0,low_memory=False)
df=df.sort_values(['season_x','name','GW'])
# df.drop(['team_x'],inplace=True,axis=1)
```

```python
all_players_seasons=[]
for player in df.name.unique():
    player_weeks= df[df.name==player]
#     team_x=df.team_x[df.name==player]
    for week in player_weeks.GW.unique():        
        last3_data= player_weeks.iloc[max(0,week-4):week].iloc[:-1]
        last10_data= player_weeks.iloc[max(0,week-10):week].iloc[:-1]
        all_last= player_weeks[:week-1]
        if last3_data.shape[0]>0:
            player_instance={'name': player,'GW':week,'season_x':last3_data.season_x.iloc[0]}
            player_instance['last3_goals']= last3_data.goals_scored.sum()
            player_instance['last3_ycards']=  last3_data.yellow_cards.sum()
            player_instance['last3_assists']= last3_data.assists.sum()            
            player_instance['ratio_goals_scored']= all_last.goals_scored.mean() 
            player_instance['ratio_assists']= all_last.assists.mean()
            player_instance['last3_ratio_points']= last3_data.total_points.sum()/min(last3_data.shape[0],3.0)
            player_instance['ratio_minutes_played']= all_last.minutes.mean()
            player_instance['ratio_saves']= all_last.saves.mean()
            player_instance['player_value']= player_weeks[player_weeks.GW==week].value.values[0]
            player_instance['ratio_own_goals']= all_last.own_goals.mean()
            player_instance['ratio_penalties_missed']= all_last.penalties_missed.mean() 
            player_instance['ratio_penalties_saved']= all_last.penalties_saved.mean()
            player_instance['ratio_selection']= all_last.selected.mean()
            player_instance['ratio_creativity']= all_last.creativity.mean()
            player_instance['ratio_threat']= all_last.threat.mean()
            all_players_seasons.append(player_instance)
```

```python
all_players_seasons=pd.DataFrame(all_players_seasons)
mod_df=pd.merge(df,all_players_seasons,how='left')
mod_df
```

```python
cols=mod_df.columns[mod_df.isna().sum()>0]
cols
```

```python
mod_df.team_a_score.fillna(0,inplace=True)
mod_df.team_h_score.fillna(0,inplace=True)
mod_df.last3_goals.fillna(0,inplace=True)
mod_df.last3_ycards.fillna(0,inplace=True)

mod_df.last3_assists.fillna(0,inplace=True)
mod_df.ratio_goals_scored.fillna(0,inplace=True)
mod_df.ratio_assists.fillna(0,inplace=True)
mod_df.last3_ratio_points.fillna(0,inplace=True)

mod_df.ratio_minutes_played.fillna(0,inplace=True)
mod_df.ratio_saves.fillna(0,inplace=True)
mod_df.player_value.fillna(0,inplace=True)
mod_df.ratio_own_goals.fillna(0,inplace=True)

mod_df.ratio_penalties_missed.fillna(0,inplace=True)
mod_df.ratio_penalties_saved.fillna(0,inplace=True)
mod_df.ratio_selection.fillna(0,inplace=True)
mod_df.ratio_creativity.fillna(0,inplace=True)

mod_df.ratio_threat.fillna(0,inplace=True)
```
